# Import the libraries

```python
# import hierarchical clustering libraries
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sn
import os
```

```python
os.chdir("/home/prathikm/Desktop/ds_assignments/in_class_notebook_and_csvs")
Univ = pd.read_csv("Universities.csv") #import universities dataset
Univ.head()
```

# Create and assign normalization function

```python
# Normalization function - this is function definition
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
# Normalized data frame (considering only the numerical part of data) for unbiased e.distance calculation
# since expenses hevily impacts the equation. Excluding Univ column as min-max cannot be applied to strings
df_norm = norm_func(Univ.iloc[:,1:])
```

# Plot the dendrogram

```python
# create dendrogram
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='single')) #single linkage method
```

# Create clusters - here we specify 4 clusters

```python

hc = AgglomerativeClustering(n_clusters=4, affinity = 'euclidean', linkage = 'single')
```

# Predict for normalized data

```python
y_hc = hc.fit_predict(df_norm)
y_hc
```

```python
#the above are cluster numbers 0 to 3 = 4
```

```python
Clusters=pd.DataFrame(y_hc,columns=['Clusters']) #just attach cluster to datafram
```
